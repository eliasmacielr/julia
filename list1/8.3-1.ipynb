```julia
using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin
        0 <= x11
        0 <= x12
        0 <= x13
        0 <= x14
        0 <= x21
        0 <= x22
        0 <= x23
        0 <= x24
        0 <= x31
        0 <= x32
        0 <= x33
        0 <= x34
        0 <= x41
        0 <= x42
        0 <= x43
        0 <= x44
end)

@constraint(m, x11 + x12 + x13 + x14 == 1)
@constraint(m, x21 + x22 + x23 + x24 == 1)
@constraint(m, x31 + x32 + x33 + x34 == 1)
@constraint(m, x41 + x42 + x43 + x44 == 1)

@constraint(m, x11 + x21 + x31 + x41 == 1)
@constraint(m, x12 + x22 + x32 + x42 == 1)
@constraint(m, x13 + x23 + x33 + x43 == 1)
@constraint(m, x14 + x24 + x34 + x44 == 1)


@objective(m, Min,
     8*x11+6*x12+5*x13+7*x14
    +6*x21+5*x22+3*x23+4*x24
    +7*x21+8*x22+4*x23+6*x24
    +6*x21+7*x22+5*x23+6*x24)

print(m)

JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x11_value=JuMP.value(x11)
x12_value=JuMP.value(x12)
x13_value=JuMP.value(x13)
x14_value=JuMP.value(x14)
x21_value=JuMP.value(x21)
x22_value=JuMP.value(x22)
x23_value=JuMP.value(x23)
x24_value=JuMP.value(x24)
x31_value=JuMP.value(x31)
x32_value=JuMP.value(x32)
x33_value=JuMP.value(x33)
x34_value=JuMP.value(x34)
x41_value=JuMP.value(x41)
x42_value=JuMP.value(x42)
x43_value=JuMP.value(x43)
x44_value=JuMP.value(x44)

println("Valor func objetivo, Min = ", obj_value)
println("x11 = ", x11_value)
println("x12 = ", x12_value)
println("x13 = ", x13_value)
println("x14 = ", x14_value)
println("x21 = ", x21_value)
println("x22 = ", x22_value)
println("x23 = ", x23_value)
println("x24 = ", x24_value)
println("x31 = ", x31_value)
println("x32 = ", x32_value)
println("x33 = ", x33_value)
println("x34 = ", x34_value)
println("x41 = ", x41_value)
println("x42 = ", x42_value)
println("x43 = ", x43_value)
println("x44 = ", x44_value)
```
