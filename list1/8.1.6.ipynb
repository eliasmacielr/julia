```julia
using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin 
        0 <= x11        # 
        0 <= x12          
        0 <= x13 
        0 <= x14 
        0 <= x21 
        0 <= x22 
        0 <= x23
        0 <= x24 
        0 <= x31 
        0 <= x32 
        0 <= x33
        0 <= x34
        0 <= w1
        0 <= w2
        0 <= w3
        0 <= w4 
        0 <= w5
end)  

@constraint(m, x11 + x12 + x13 + x14 + w1 == 60)        #
@constraint(m, x21 + x22 + x23 + x24 + w2 == 80) 
@constraint(m, x31 + x32 + x33 + x34 + w3 == 40) 
@constraint(m, x11 + x21 + x31 == 40) 
@constraint(m, x12 + x22 + x32 == 60) 
@constraint(m, x13 + x23 + x33 - w4 == 20) 
@constraint(m, x14 + x24 + x34 + w5 == 60) 
@objective(m, Max, 800*x11 + 700*x12 + 500*x13 + 200*x14 + 500*x21 + 200*x22 + 100*x23 + 300*x24 + 
    600*x31 + 400*x32 + 300*x33 + 500*x34)           # max ganancia

print(m)
JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x11_value=JuMP.value(x11)
x12_value=JuMP.value(x12)
x13_value=JuMP.value(x13)
x14_value=JuMP.value(x14)
x21_value=JuMP.value(x21)
x22_value=JuMP.value(x22)
x23_value=JuMP.value(x23)
x24_value=JuMP.value(x24)
x31_value=JuMP.value(x31)
x32_value=JuMP.value(x32)
x33_value=JuMP.value(x33)
x34_value=JuMP.value(x34)
w1_value=JuMP.value(w1)
w2_value=JuMP.value(w2)
w3_value=JuMP.value(w3)
w4_value=JuMP.value(w4)

println("Valor func objetivo, Ganancia=", obj_value)
println("x11 = ", x11_value)
println("x12 = ", x12_value)
println("x13 = ", x13_value)
println("x14 = ", x14_value)
println("x21 = ", x21_value)
println("x22 = ", x22_value)
println("x23 = ", x23_value)
println("x24 = ", x24_value)
println("x31 = ", x31_value)
println("x32 = ", x32_value)
println("x33 = ", x33_value)
println("x34 = ", x34_value)
println("x34 = ", w1_value)
println("x34 = ", w2_value)
println("x34 = ", w3_value)
println("x34 = ", w4_value)
```
