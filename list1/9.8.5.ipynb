```julia
using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin
        0 <= xA <= 4
        0 <= xB <= 3
        0 <= xC <= 5
        0 <= xD <= 3
        0 <= xE <= 5
        0 <= xF <= 7
        0 <= xG <= 2
        0 <= xH <= 3
        0 <= xI <= 4
        0 <= xJ <= 2
        
        0 <= yC
        0 <= yD
        0 <= yE
        0 <= yF
        0 <= yG
        0 <= yH
        0 <= yI
        0 <= yJ
        0 <= yFIN
end)

@constraint(m, yC >= 32 - xA)
@constraint(m, yJ >= yC + 36 - xC)
@constraint(m, yF >= 28 - xB)
@constraint(m, yE >= 28 - xB)
@constraint(m, yD >= 28 - xB)
@constraint(m, yJ >= yF + 54 - xF)
@constraint(m, yG >= yD + 16 - xD)
@constraint(m, yH >= yG + 17 - xG)
@constraint(m, yH >= yE + 32 - xE)
@constraint(m, yI >= yE + 32 - xE)
@constraint(m, yI >= yG + 17 - xG)
@constraint(m, yFIN >= yI + 34 - xI)
@constraint(m, yFIN >= yH + 20 - xH)
@constraint(m, yFIN >= yJ + 18 - xJ)


@constraint(m, yFIN <= 70)

@objective(m, Min, 5000000*xA + 7000000xB + 8000000*xC + 4000000*xD + 5000000*xE + 6000000*xF + 3000000xG + 4000000*xH 
    + 9000000*xI + 2000000*xJ)

print(m)

JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
xA_value=JuMP.value(xA)
xB_value=JuMP.value(xB)
xC_value=JuMP.value(xC)
xD_value=JuMP.value(xD)
xE_value=JuMP.value(xE)
xF_value=JuMP.value(xF)
xG_value=JuMP.value(xG)
xH_value=JuMP.value(xH)
xI_value=JuMP.value(xI)
xJ_value=JuMP.value(xJ)


yC_value=JuMP.value(yC)
yD_value=JuMP.value(yD)
yE_value=JuMP.value(yE)
yF_value=JuMP.value(yF)
yG_value=JuMP.value(yG)
yH_value=JuMP.value(yH)
yI_value=JuMP.value(yI)
yJ_value=JuMP.value(yJ)
yFIN_value=JuMP.value(yFIN)

println("Valor func objetivo, Min = ", obj_value)
println("xA = ", xA_value)
println("xB = ", xB_value)
println("xC = ", xC_value)
println("xD = ", xD_value)
println("xE = ", xE_value)
println("xF = ", xF_value)
println("xG = ", xG_value)
println("xH = ", xH_value)
println("xI = ", xI_value)
println("xJ = ", xJ_value)
```
